#Part 1

```python
import json
import random
import nltk
import numpy as np
import pickle
from nltk.stem import WordNetLemmatizer
```

```python
path = '/content/intents.json'
```

```python
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.optimizers import SGD
```

```python
lemmatizer = WordNetLemmatizer()
intents = json.loads(open(path).read())
```

```python
words = []
classes = []
documents = []
ignore_letters = ['?','!','.',',']
```

```python
for inte in intents['intents']:
  for pattern in inte['patterns']:
    tokenized_word = nltk.word_tokenize(pattern)
    words.extend(tokenized_word)
    documents.append((tokenized_word,inte['tag']))
    if inte['tag'] not in classes:
      classes.append(inte['tag'])
print(documents)
```

```python
words = [lemmatizer.lemmatize(i) for i in words if i not in ignore_letters ]
words = sorted(set(words))
print(words)
```

```python
classes = sorted(set(classes))
print(classes)
```

```python
pickle.dump(classes,open('classes.pkl','wb'))
pickle.dump(words,open('words.pkl','wb'))
```

```python
training  = []
output_empty = [0] * len(classes)

for document in documents:
    bag = []
    word_patterns = document[0]
    word_patterns = [lemmatizer.lemmatize(word.lower()) for word in word_patterns]
    for word in words:
        bag.append(1) if word in word_patterns else bag.append(0)
    output_row = list(output_empty)
    output_row[classes.index(document[1])] = 1
    training.append([bag, output_row])
```

```python
random.shuffle(training)
training = np.array(training)

train_x = list(training[:, 0])
train_y = list(training[:, 1])
```

```python
train_x
```

```python
model = Sequential()

model.add(Dense(128, input_shape=(len(train_x[0]),),activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))
```

```python
sgd = SGD(lr=0.01,  momentum = 0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
hist = model.fit(np.array(train_x), np.array(train_y), epochs=200, batch_size=5, verbose=1)
model.save('Movie.h5',hist)
```

#Part 2

```python
import random
import json
import pickle
import numpy as np
```

```python
import nltk
nltk.download('omw-1.4')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

from tensorflow.keras.models import load_model
```

```python
lemmatizer = WordNetLemmatizer()
path = '/content/intents.json'
intents = json.loads(open(path).read())

words = pickle.load(open('words.pkl','rb'))
classes = pickle.load(open('classes.pkl','rb'))

model = load_model('Movie.h5')
```

```python
def clean_up_sentence(sentence):
    sentence_words = nltk.word_tokenize(sentence)
    sentence_words = [lemmatizer.lemmatize(word) for word in sentence_words]
    return sentence_words
```

```python
def bag_of_words(sentence):
    sentence_words = clean_up_sentence(sentence)
    bag = [0] * len(words)
    for w in sentence_words:
        for i, word in enumerate(words):
            if word == w:
                bag[i] = 1
    return np.array(bag)
```

```python
def predict_class(sentence):
    bow = bag_of_words(sentence)
    res = model.predict(np.array([bow]))[0]
    ERROR_THRESHOLD = 0.25
    results = [[i,r] for i, r in enumerate(res) if r>ERROR_THRESHOLD]

    results.sort(key=lambda x:x[1], reverse=True)

    return_list = []

    for r in results:
        return_list.append({'intent':classes[r[0]], 'probability':str(r[1])})
    return return_list
```

```python
def get_response(intents_list, intents_json):
    tag = intents_list[0]['intent']
    list_of_intents = intents_json['intents']
    for i in list_of_intents:
        if i['tag'] == tag:
            result = random.choice(i['responses'])
            break
    return result
```

```python
while True:
    message = input("")
    if(message == "stop"):
      print("Bye Bye..")
      break
    ints = predict_class(message)
    res = get_response(ints, intents)
    print(res)
```
